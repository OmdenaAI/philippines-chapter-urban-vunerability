```python
import pandas as pd
import numpy as np
import os
from pathlib import Path
```

```python
# Setting up directories

WORKINGDIR = Path(os.getcwd())
PROJECTROOT = WORKINGDIR.parents[1]

DTI_DATA_DIR = Path("data", "dti-data")
```

```python
dti_data_filenames = os.listdir(Path(PROJECTROOT,DTI_DATA_DIR))
dti_data_filenames
```

```python
sample_df = pd.read_csv(Path(PROJECTROOT,DTI_DATA_DIR,dti_data_filenames[0]))
sample_df
```

```python
YEARS = ["2018", "2019", "2020", "2021", "2022"]
P_I = list(sample_df["Pillar / Indicator"].unique())
```

```python
combinations_list = []
combinations_appended = []
```

```python
for i in YEARS:
    for j in P_I:
        list_ = [i,j]
        combinations_list.append(list_)
```

```python
row_list = []

for csv_file in dti_data_filenames:
    df = pd.read_csv(Path(PROJECTROOT,DTI_DATA_DIR,csv_file))
    
    dict_ = {}
    
    dict_["city_municipality"] = df["City_Municipality"].iloc[0]

    for i in range(len(combinations_list)):
        value = df[combinations_list[i][0]].loc[df["Pillar / Indicator"]==combinations_list[i][1]]
        feature = combinations_list[i][0] + " - " + combinations_list[i][1]
        
        try:
            value = float(value)
        except ValueError:
            value = np.nan
            
        dict_[feature] = value
    
    row_list.append(dict_)
```

```python
merged_df = pd.DataFrame.from_dict(row_list)
```

```python
merged_df.head()
```

```python
drop_cols = []

for col in merged_df.columns:
    null_perc = merged_df[col].isnull().sum()*100/len(merged_df)
    if null_perc > 60:
        drop_cols.append(col)
```

```python
merged_df.drop(columns=drop_cols, inplace=True)
```

```python
merged_df.sort_values("city_municipality").to_csv(Path(PROJECTROOT,"data", "dti-data-merged", "merged-dti-data.csv"), index=False)
```
