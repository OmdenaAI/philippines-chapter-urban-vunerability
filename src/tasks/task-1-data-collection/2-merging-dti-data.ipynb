```python
sample_df = pd.read_csv(Path(PROJECTROOT,DTI_DATA_DIR,dti_data_filenames[0]))
sample_df
```

```python
YEARS = ["2018", "2019", "2020", "2021", "2022"]
P_I = list(sample_df["Pillar / Indicator"].unique())
```

```python
combinations_list = []
combinations_appended = []
```

```python
for i in YEARS:
    for j in P_I:
        list_ = [i,j]
        combinations_list.append(list_)
        
        column_name = i + " - " + j
        combinations_appended.append(column_name)
```
