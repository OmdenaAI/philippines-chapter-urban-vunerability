```python
# Import Required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Read and Inspect The Dataset

```python
dti = pd.read_csv("merged-dti-data.csv")
dti.head()
```

```python
cols = dti.columns[dti.columns.str.startswith('2021')].tolist()
dti_2021 = dti[['city_municipality'] + cols]
dti_2021.head()
```

```python
dti_2021.shape
```

```python
dti_2021 = pd.DataFrame(dti_2021).set_index('city_municipality')
```

```python
# Finding the rows that has 100% null values
dti_2021[dti_2021.isna().any(axis=1)]
```

There are 27 municipalities which don't have any data for the year 2021. So let's remove those municipalities.

```python
# Remove rows having 100% null values
dti_2021 = dti_2021.dropna(how='all', axis=0)
```

```python
dti_2021.shape
```

```python
print(round(dti_2021[dti_2021 == 0].count(axis=0)*100/len(dti_2021.index)).sort_values())
```

Financial Technology Capacity, Presence of Business and Professional Organizations, Accomodation Capacity, Recognition Performance have values as zero for more than 25% of municipalities.

```python
# Clubbing indicators to different pillars such as economic dynamism, government efficiency, infrastructure, resiliency
economic_dynamism = dti_2021.columns[:10]
government_efficiency = dti_2021.columns[10:20]
infrastructure = dti_2021.columns[20:30]
resiliency = dti_2021.columns[30:40]
```
