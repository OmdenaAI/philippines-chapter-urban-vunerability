## EDA of DTI Data 2021

```python
dti_data_2021 = df.iloc[:, 128:168]
```

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
import scipy.stats as st
from sklearn import ensemble, tree, linear_model
import missingno as msno
```

```python
dti_data_2021.head()
```

```python
dti_data_2021[dti_data_2021.isna().any(axis=1)]
```

```python
dti_data_2021 = dti_data_2021.dropna(how='all', axis=0)
```

```python
dti_data_2021.shape
```

```python
dti_data_2021.columns.tolist()
```

```python
correlation = dti_data_2021.corr()
k= 11
cols = correlation.nlargest(k,'2021 - Hea')['2021 - Hea'].index
print(cols)
```

```python
cm = np.corrcoef(dti_data_2021[cols].values.T)
f , ax = plt.subplots(figsize = (7,6))
sns.heatmap(cm, vmax=.8, linewidths=0.01,square=True,annot=True,cmap='viridis',linecolor="white",xticklabels = cols.values ,annot_kws = {'size':12},yticklabels = cols.values)
```

```python
df_merged = pd.merge(dti_data_2021, df1, left_index= True, right_index= True)
```

```python
correlation = df_merged.corr()
k= 13
cols = correlation.nlargest(k,'Total-health-fac-2021')['Total-health-fac-2021'].index
print(cols)
```

```python
cm = np.corrcoef(df_merged[cols].values.T)
f , ax = plt.subplots(figsize = (10,12))
sns.heatmap(cm, vmax=.8, linewidths=0.01,square=True,annot=True,cmap='viridis',linecolor="white",xticklabels = cols.values ,annot_kws = {'size':12},yticklabels = cols.values)
```
